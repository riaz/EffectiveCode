```python
squad_en_dataset = load_dataset("json", field="data") # we can also apply delimiters .
```

```python
squad_en_dataset
```

```python
# shuffling the data
rand_train_sample = squad_en_dataset["train"].shuffle(seed=42).select(range(20)) # randomly sample 20 examples
rand_train_sample[1].keys()
```

```python
example_1 = rand_train_sample[0]
```

```python
def display(resp: str, span=10) -> str:
    ans = []
    for x in range(0, len(resp), span):
        ans.append(resp[x:x+span])
    return "\n".join(ans)
```

```python
#example_1["paragraphs"]
for x in example_1["paragraphs"]:
    print("Context\n\n")
    print(display(x['context'], span=100))
    for q in x['qas']:
        print(f"\nQuestion : {q['question']}\n")
        for ans in q['answers']:
            resp = x['context'][ans['answer_start']:]            
            print("\n"+ display(resp, span=100)+ "\n")
            print(f"\nDirect Answer: {ans['text']}\n")
```
